```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# =====================================================
# 1. LOAD EXCEL
# =====================================================
excel_path = "D:/GeneAgent3/干净数据/20251118_summary_results_all_models.ratio.80.xlsx"
xls = pd.ExcelFile(excel_path)

# =====================================================
# 2. SETUP STORAGE
# =====================================================
models = []
overall = []

fields = ["symbol_match", "gene_name_match", "disease_match", "phenotype_match"]
per_field = {f: [] for f in fields}

# =====================================================
# 3. READ EACH MODEL SHEET
# =====================================================
for sheet in xls.sheet_names:
    df = pd.read_excel(xls, sheet_name=sheet)
    models.append(sheet)

    # overall accuracy
    total_true = df["true_count"].sum()
    total_valid = df["valid_count"].sum()
    overall_acc = total_true / total_valid
    overall.append(overall_acc)

    # per-field
    for f in fields:
        val = float(df.loc[df["field"] == f, "match_pct"])
        per_field[f].append(val)

# =====================================================
# 4. EXACT COLOR PALETTE FROM YOUR FIGURE
# =====================================================
palette = [
    "#7BB662",  # green (Canada)
    "#F4A7A1",  # France light pink
    "#4A6FE3",  # Germany blue
    "#A085C8",  # Italy purple
    "#F4C2C2",  # Japan pink
    "#E41A1C",  # UK red
    "#F9C97A"   # US orange
]

# =====================================================
# 5. PLOT – OVERALL ACCURACY WITH LABELS
# =====================================================
plt.figure(figsize=(16, 6))
bars = plt.bar(models, overall, color=palette[:len(models)])

# add numbers on top
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 0.01,
        f"{height:.3f}",
        ha='center',
        va='bottom',
        fontsize=12
    )

plt.ylim(0.75, 1.02)
plt.ylabel("Accuracy")
plt.title("Overall Accuracy per Model (Styled Palette)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# =====================================================
# 6. PLOT – PER-FIELD GROUPED BAR WITH LABELS
# =====================================================
x = np.arange(len(models))
width = 0.18

plt.figure(figsize=(18, 7))

for i, f in enumerate(fields):
    bars = plt.bar(
        x + i*width,
        per_field[f],
        width=width,
        color=palette[i],
        label=f.replace("_match", "").title()
    )

    # add value labels
    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width()/2,
            height + 0.01,
            f"{height:.3f}",
            ha='center',
            va='bottom',
            fontsize=10
        )

plt.xticks(x + width, models, rotation=45, ha='right')
plt.ylim(0.75, 1.02)
plt.ylabel("Accuracy")
plt.title("Per-field Accuracy per Model (Styled Palette)")
plt.legend()
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors

# === Your main pastel palette ===
palette_main = [
    "#7AB669",  # green
    "#F4B8B8",  # pink
    "#4365E2",  # blue
    "#A58ACB",  # purple
    "#F5CDCD",  # light pink
    "#E21620",  # red
    "#F7C66A"   # yellow
]

# === Make darker variants for bottom stacked bars ===
def darken(color, factor=0.55):
    c = np.array(mcolors.to_rgb(color))
    return (c * factor).clip(0,1)

palette_bottom = [darken(c) for c in palette_main]

# === Data (same as your screenshot) ===
models = ["GPT-4o","Claude-Opus-4","DeepSeek","Grok-4","Qwen-3.5","Gemini-2.5","GPT-5"]
bottom_vals = np.array([1511,1536,1146,1451,1151,1595,0])
db_gap      = np.array([  35,  19,  22,  16,  24,  22,0])
top_vals    = np.array([1136,1288,1123, 751,1080, 831,0])

# === Plotting ===
fig, ax = plt.subplots(figsize=(14,6))
x = np.arange(len(models))

gap_scale = 8
db_gap_scaled = db_gap * gap_scale

for i in range(len(models)):
    # Bottom (dark)
    ax.bar(x[i], bottom_vals[i],
           color=palette_bottom[i], edgecolor="black")

    # DB gap (yellow, scaled)
    ax.bar(x[i], db_gap_scaled[i],
           bottom=bottom_vals[i],
           color="#F7C66A", edgecolor="black")

    # Top (pastel)
    ax.bar(x[i], top_vals[i],
           bottom=bottom_vals[i] + db_gap_scaled[i],
           color=palette_main[i], edgecolor="black")

    # === Correct text labels ===
    # Bottom
    if bottom_vals[i] > 0:
        ax.text(x[i], bottom_vals[i]/2, bottom_vals[i],
                ha="center", va="center", color="white", fontsize=14)

    # Gap: display original value, centered in scaled height
    if db_gap[i] > 0:
        ax.text(x[i],
                bottom_vals[i] + db_gap_scaled[i] / 2,
                str(db_gap[i]),  # show real gap
                ha="center", va="center", color="black", fontsize=14)

    # Top
    if top_vals[i] > 0:
        ax.text(x[i],
                bottom_vals[i] + db_gap_scaled[i] + top_vals[i] / 2,
                top_vals[i],
                ha="center", va="center", color="black", fontsize=14)

ax.set_xticks(x)
ax.set_xticklabels(models, rotation=30)
ax.set_ylabel("Count")
ax.set_title("Phenotype Prediction Classification (Stacked DB Gap)")

plt.tight_layout()
plt.show()

```
